```python
# we can use membership operators
my_numbers = [10,20,30,40]
your_numbers = (10,20,30,40)
name = "Python"
```

```python
10 in my_numbers
```

```python
10 in your_numbers
```

```python
15 in my_numbers
```

```python
15 not in my_numbers
```

```python
# iterating a sequence with while

index = 0
while index < len(my_numbers):
    print(my_numbers[index])
    index = index + 1
```

```python
# iterating a sequence with for - for each
for my_number in my_numbers:
    print(my_number)
```
