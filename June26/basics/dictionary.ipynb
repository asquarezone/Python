```python
# dictionary has name and value (key value)
student = {
    # key   #value
    "name": "xyz",
    "age": 21,
    "courses": ["python", "genai", "devops"]
}
```

```python
student['courses']
```

```python
student['name']
```

```python
# get all keys
list(student.keys())
```

```python
#get all values
list(student.values())
```

```python
# iterating a dictionary
for key in student:
    print(key)
```

```python
for key in student:
    print(f"{key} => {student[key]}")
```

```python
for key, value in student.items():
    print(f"{key} => {value}")
```
