## Map, Reduce and Filter

````python
numbers = [1,2,3,4,5]
````

````python
def plus_one(number):
    return number + 1
````

````python
list(map(plus_one, numbers))
````

````python
numbers = [1,2,3,4,5]
def square(number):
    return number * number

list(map(square, numbers))
````

````python
list(map(lambda x: x * x, numbers))
````

````python
from functools import reduce
numbers = [1,2,3,4,5]
# sum of all items in a sequence
print(reduce(lambda x,y : x + y, numbers))
# product of all items in sequence
print(reduce(lambda x,y : x * y, numbers))
````

````python
def is_prime(number):
    """Checks if the number is prime or not

    Args:
        number (_type_): number 

    Returns:
        _type_: Returns True if prime False otherwise
    """
    if number < 2:
        return False
    is_prime_number= True
    index = 2
    while index <= number // 2:
        if number % index == 0:
            is_prime_number = False
            break
        index += 1
    return is_prime_number
````

````python
is_prime(7)
````

````python
numbers = list(range(1, 100))
````

````python
list(filter(is_prime,numbers))
````

## List comprehension

* syntax
```python
[ <expression> for <item > in <items> <conditional> ]
```

````python
[ x for x in range(1,10)]
````

````python
[ x for x in range(1,10) if x % 2 == 0]
````

````python
numbers = [1,2,3,4,5]
````

````python
# increment each value by 1
# map
print(list(map(lambda x: x + 1, numbers)))
````

````python
# list comprehension
[ number + 1 for number in numbers]
````

````python
# square of all numbers
[ number * number for number in numbers]
````

````python
# square of all even numbers
[ number * number for number in numbers if number % 2 == 0 ]
````

````python
# raw 
new_numbers = list()
for number in numbers:
    if number % 2 == 0:
        new_numbers.append(number * number)
new_numbers
````

````python
# list of all prime numbers between 1000 and 1050
prime_numbers = []
for number in range(1000, 1050):
    if is_prime(number):
        prime_numbers.append(number)

prime_numbers
````

````python
[ number for number in range(1000,1050) if is_prime(number)]
````

````python
list(filter(is_prime, range(1000, 1050)))
````
