````python
# i want to print all the numbers which are divisible by 3 or 5 till 20
index = 1
while index < 20:
    if index % 3 == 0 or index % 5 == 0:
        print(index)
    index += 1

print("Completed")
````

````python
# is a number prime or not
number = int(input("Enter the number"))
is_prime = True
index = 2
while index < number:
    if number % index == 0:
        is_prime = False
        # i want to come out of this loop
        break
    index += 1

if is_prime:
    print(f"{number} is prime")
else:
    print(f"{number} is not prime")
````

````python
# Give me sum of all numbers apart from numbers divisible by 3 till 100
index = 0
sum = 0
while index <= 10:
    index += 1
    if  index % 3 == 0:
        continue
    sum += index

print(sum)
````
