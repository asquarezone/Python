```python
class Student:
    # class member
    institute = "Quality Thought"

    
    def __init__(self, name, age):
        # instance member
        # self is object
        self.name = name
        self.age = age
```

```python
class QTStudent:

    institute = "Quality Thought"
    
    def __init__(qt, name, age):
        qt.name = name
        qt.age = age
```

```python
s1 = Student(name="abc", age=21)
s2 = Student(name="abd", age=21)
```

```python
s2.age
```

```python
class XStudent:
    # class member
    institute = "Quality Thought"

    
    def __init__(this, name, age):
        # instance member
        # self is object
        this.name = name
        this.age = age
```

```python
xs1 = XStudent(name="abc", age=21)
xs1.name
```

```python
class Student:
    school = "QT"

    def __init__(self, name, age, courses=[]):
        self.name = name
        self.age = age
        self.courses = courses

    # instance method
    def print_details(self):
        print(f"Name is {self.name} and age is {self.age}")
```

```python
s1 = Student("name1", 21)
s2 = Student("name2", 22)

# calling an instance method
s1.print_details()
# the above syntax is what we use the most

# other way of calling instance method
Student.print_details(s1)
```

```python
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def print_details(self):
        print(f"Make is {self.make} and model is {self.model}")
```

```python
# Car is derived from Vehicle
# Vehicle => Parent
# Car => Child
class Car(Vehicle):
    pass
```

```python
c1 = Car(make="Hyundai", model="i20")
```

```python
c1.print_details()
```

```python
class Truck(Vehicle):
    
    def print_details(self):
        print(f"Truck Make is {self.make} and model is {self.model}")
```

```python
t1 = Truck(make="Tata", model="Sumo")
t1.print_details()
```
