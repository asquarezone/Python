```python
def print_numbers(start, finish):
    """This function prints the numbers from start to finish

    Args:
        start (_type_): start
        finish (_type_): finish
    """
    while start <= finish:
        print(start)
        start += 1
```

```python
print_numbers(1, 5)
```

```python
def is_even_number(number):
    """This function will check if the number is even or not

    Args:
        number (_type_): number

    Returns:
        _type_: Returns True if even False otherwise
    """
    return number % 2 == 0
```

```python
is_even_number(7)
```

```python
is_even_number(70)
```

```python
def is_prime(number):
    """Checks if the number is prime or not

    Args:
        number (_type_): number 

    Returns:
        _type_: Returns True if prime False otherwise
    """
    if number < 2:
        return False
    is_prime_number= True
    index = 2
    while index <= number // 2:
        if number % index == 0:
            is_prime_number = False
            break
        index += 1
    return is_prime_number
```

```python
is_prime(7)
```

```python
is_prime(70)
```

```python
is_prime(0)
```

```python
is_prime(-7)
```
