## Euler - 1

* [Refer Here](https://projecteuler.net/problem=1)

```python
index = 1
max = 1000
sum = 0
while index < max:
    if index % 3 == 0 or index % 5 == 0:
        sum += index
    index += 1
print(sum)
```

## Euler - 2

* [Refer Here](https://projecteuler.net/problem=2)

```python
first = 1
second = 2
sum = 2
max = 4000000
while first + second < max:
    third = first + second
    if third % 2 == 0:
        sum += third
    first = second
    second = third

print(sum)
```

## Euler - 3

* [Refer Here](https://projecteuler.net/problem=3)

```python
# First lets find all factors of a number
number = 20
index = 2
while index < number:
    if number % index == 0:
        print(index)
    index += 1
```

```python
# Lets find the factors from largest to smallest
number = 20
index = number - 1
while index > 1:
    if number % index == 0:
        print(index)
    index -= 1
```

```python
# largest factor
number = 20
index = number - 1
while index > 1:
    if number % index == 0:
        print(index)
        break
    index -= 1
```

```python
# finding if the number is prime 
# number is stored in index
# if index is prime
index = 70
prime_index = 2
is_prime = True
while prime_index < index:
    if index % prime_index == 0:
        is_prime = False
        break
    prime_index += 1
if is_prime:
    print("prime")
else:
    print("not prime")
```

```python
# largest factor
number = 13195
index = number // 2
while index > 1:
    if number % index == 0:
        prime_index = 2
        is_prime = True
        while prime_index < index // 2:
            if index % prime_index == 0:
                is_prime = False
                break
            prime_index += 1
        if is_prime:
            print(index)
            break
    index -= 1
```
