## Euler - 2

```python
first = 1
second = 2
sum = 2
max = 4000000
while first + second < max:
    third = first + second
    if third % 2 == 0:
        sum += third
    first = second
    second = third

print(sum)
```

```python
# Euler 1 functions

def is_divisible_by_3_or_5(number):
    """This function checks if the number is divisible by 3 or 5

    Args:
        number (_type_): number to be checked

    Returns:
       True if divisible by 3 or 5 False otherwise
    """
    return number % 3 == 0 or number % 5 == 0
```

```python
# Euler 1
number = 1000
index = 1
sum = 0
while index < number:
    if is_divisible_by_3_or_5(index):
        sum += index
    index += 1

print(f"Result is {sum}")
```

```python
def is_factor_of(number, index):
    """Checks whether index is factor of number 

    Returns True if factor False otherwise
    """
    return number % index == 0

def is_prime_number(number):
    """Checks if the number is prime or not

    Returns:
      True if prime False otherwise
    """
    if number < 2:
        return False
    is_prime = True
    index = 2
    while index <= number // 2:
        if is_factor_of(number, index):
            is_prime = False
            break
        index += 1
    return is_prime
    
```

```python
def is_prime_factor(number, index):
    """Returns True if prime factor False otherwise
    """
    return is_factor_of(number, index) and is_prime_number(index)
```

```python
number = 13195
index = number // 2
result = 0
while index > 2:
    if is_prime_factor(number, index):
        result = index
        break
    index -= 1
print(f"The largest factor of {number} is {result}")
```

```python
def is_pallendrome(number):
    """Checks if the number is pallendrome or not

    Args:
        number (_type_): number

    Returns:
       True if pallendrome False otherwise
    """
    return str(number) == str(number)[::-1]
```

```python
x = 999
y = 999
largest_pallendrome = 0
result_x = 0
result_y = 0
while x >= 100:
    while y >= 100:
        number = x * y
        if is_pallendrome(number) and number > largest_pallendrome:
            largest_pallendrome = number
            result_x = x
            result_y = y
        y -= 1
    x -= 1
print(f"largest pallendrom is {largest_pallendrome} => {result_x} * {result_y}")
```
