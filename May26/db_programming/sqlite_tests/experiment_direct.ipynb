```python
# standard library
import sqlite3
```

```python
# create a connection
connection = sqlite3.connect("./data/students.db")
```

```python
# we need to create  cursor object to execute queries
cursor = connection.cursor()
```

```python
# Using cursor we can execute sql statements

CREATE_STUDENTS_QUERY = """CREATE TABLE students (
    student_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    date_of_birth TEXT,
    class_name TEXT NOT NULL,
    gender TEXT,
    phone TEXT
);
"""

cursor.execute(CREATE_STUDENTS_QUERY)
connection.commit()
connection.close()
```

```python
# ideal way of doing it 
with sqlite3.connect("./data/students.db") as connection:
    cursor = connection.cursor()
    CREATE_STUDENTS_QUERY = """CREATE TABLE students (
        student_id INTEGER PRIMARY KEY,
        first_name TEXT NOT NULL,
        last_name TEXT NOT NULL,
        date_of_birth TEXT,
        class_name TEXT NOT NULL,
        gender TEXT,
        phone TEXT
    );
    """

    cursor.execute(CREATE_STUDENTS_QUERY)
    connection.commit()
```

```python
with sqlite3.connect("./data/students.db") as connection:
    cursor = connection.cursor()
    INSERT_STATIC_QUERY = "INSERT INTO students (first_name, last_name, date_of_birth, class_name, gender, phone) VALUES ('Aarav', 'Sharma', '2010-05-12', '8-A', 'Male', '9876543210');"
    cursor.execute(INSERT_STATIC_QUERY)
    connection.commit()
```

```python
# Dynamic insertion
with sqlite3.connect("./data/students.db") as connection:
    cursor = connection.cursor()
    INSERT_QUERY = """
    INSERT INTO students
    (first_name, last_name, date_of_birth, class_name, gender, phone)
    VALUES (?, ?, ?, ?, ?, ?);
    """
    VALUES = ('Rahul', 'Sharma', '2010-06-12', '8-B', 'Male', '9876543211')
    cursor.execute(INSERT_QUERY, VALUES)
    connection.commit()
```
