```python
import time
def task(name:str="test") -> None:
    """Prints the name"""
    # simulating a db call or a network call
    time.sleep(2)
    print(f"Hello {name}!")
```

```python
import timeit
t = timeit.timeit(task, number=2)
print(f"Function time: {t}")
```

```python
import threading
def with_threads():
    t1 = threading.Thread(target=task, args=("Siddhu",))
    t2 = threading.Thread(target=task, args=("Madhu",))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
```

```python
t = timeit.timeit(with_threads, number=1)
print(f"Function time: {t}")
```
