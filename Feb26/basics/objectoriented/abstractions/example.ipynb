```python
from abc import ABC, abstractmethod

class PaymentBase(ABC):
    @abstractmethod
    def pay(self, amount:float):
        """The Pay method which defines
        the payment contract

        Args:
            amount (float): Amount to be paid
        """
```

```python
class CashPayment(PaymentBase):
    pass
```

```python
cp = CashPayment()
```

```python
class CashPayment(PaymentBase):
    def pay(self, amount:float):
        """The Pay method implements the pay contract

        Args:
            amount (float): Amount to be paid
        """
        print(f"Paid in cash the amount of {amount}")
```

```python
cp = CashPayment()
cp.pay(1000)
```
