```python
from abc import ABC, abstractmethod

class PaymentBase(ABC):
    @abstractmethod
    def pay(self, amount:float):
        """The Pay method which defines
        the payment contract

        Args:
            amount (float): Amount to be paid
        """
```

```python
class CashPayment(PaymentBase):
    pass
```

```python
cp = CashPayment()
```

```python
class CashPayment(PaymentBase):
    def pay(self, amount:float):
        """The Pay method implements the pay contract

        Args:
            amount (float): Amount to be paid
        """
        print(f"Paid in cash the amount of {amount}")
```

```python
cp = CashPayment()
cp.pay(1000)
```

# Traditional Way (OOP)

```python
class Vehicle:
    def start(self):
        pass
```

```python
class Car(Vehicle):
    def start(self):
        print("Car is starting")


class Bike(Vehicle):
    def start(self):
        print("Bike is starting")
```

```python
def start_vehicle(vehicle:Vehicle):
    vehicle.start()
```

```python
start_vehicle(Bike())
```

```python
start_vehicle(Car())
```

```python
class TV:
    def start(self):
        print("TV is starting")
```

```python
start_vehicle(TV())
```

```python
class Human:
    def run(self):
        print("human is running")

class SuperHuman(Human):
    def run(self):
        print("Super human is running in sonic speed")
```

```python
def check_run(human):
    human.run()
```

```python
check_run(Human())
```

```python
check_run(SuperHuman())
```

```python
class Horse:
    def run(self):
        print("Horse is running")
```

```python
check_run(Horse())
```
