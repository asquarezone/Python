```python
# we can add syntantic sugar to make attributes 
# and method behave like private
class Car:
    def __init__(self, make:str, model:str):
        self.__make = make
        self.__model = model

    def set_make(self, make):
        self.__make = make

    def __set_model(self, model):
        self.__model = model
```

```python
harrier = Car('Tata', 'harrier')
```

```python
harrier.__make
```

```python
harrier.set_make('Tata')
```

```python
harrier.__set_model('Harrier')
```

```python
class Car:
    def __init__(self, make, model, owner):
        self.__make = make
        self.__model = model
        self.__owner = owner

    def get_make(self):
        return self.__make
    
    def get_model(self):
        return self.__model

    def get_owner(self):
        return self.__owner
    
    def set_owner(self, owner):
        self.__owner = owner
```

```python
car1 = Car("Toyota", "Camry", "Khaja")
car2 = Car("Honda", "Civic", "Rahul")
car3 = Car("Tesla", "Model 3", "Anita")
print(car1.get_make())   # Toyota
print(car1.get_model())  # Camry
print(car1.get_owner())  # Khaja
print(car2.get_make())   # Honda
print(car2.get_model())  # Civic
print(car2.get_owner())  # Rahul
car1.set_owner("Arjun")
print(car1.get_owner())  # Arjun
```

```python
# With property decorator
class Car:
    def __init__(self, make, model, owner):
        self.__make = make
        self.__model = model
        self.__owner = owner

    # getters
    @property
    def make(self):
        return self.__make
    
    @property
    def model(self):
        return self.__model
    
    @property
    def owner(self):
        return self.__owner
    
    @owner.setter
    def owner(self, owner):
        self.__owner = owner

```

```python
car1 = Car("Toyota", "Camry", "Khaja")
print(f"make = {car1.make}\nmodel = {car1.model}\nowner = {car1.owner}")
car1.owner = ".........."
print(f"make = {car1.make}\nmodel = {car1.model}\nowner = {car1.owner}")
```

```python
class BankAcount:
    def __init__(self, name:str, mobile_no, email_id, aadhar_id):
        self.__name:str = name
        self.__mobile_no = mobile_no
        self.__email_id = email_id
        self.__aadhar_id = aadhar_id

    @property
    def name(self) -> str:
        return self.__name
    
    @property
    def mobile_no(self):
        return self.__mobile_no
    
    @mobile_no.setter
    def mobile_no(self, mobile_no):
        self.__mobile_no = mobile_no
    
    @property
    def email_id(self):
        return self.__email_id
    
    @email_id.setter
    def email_id(self, email_id):
        self.__email_id = email_id
    
    @property
    def aadhar_id(self):
        return self.__aadhar_id

```

```python
ram_bank_account = BankAcount("Ram", 9876543210, "ram@gmail.com", 123456789012)
```

```python
ram_bank_account.mobile_no
```

```python
ram_bank_account.mobile_no = "9876543211"
```
