````python
# members
class NewBankAccount:
    # class level member
    bank_name: str = "SBI"

    # constructor (initializer)
    def __init__(self, balance:float, number:str, name:str):
        self.balance = balance
        self.account_number = number
        self.customer_name = name

    def withdraw(self, amount: float):
        self.balance -= amount
````

````python
user4 = NewBankAccount(balance=1000, number="99999999", name="abc")
print(user4.account_number)
print(user4.balance)
print(user4.customer_name)
````

````python
user4.withdraw(50)
print(user4.balance)
````
