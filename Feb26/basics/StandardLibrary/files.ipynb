```python
# context manager syntax

with open('./data/1.txt', 'r') as file:
    print(file.readlines())
```

```python
# write a file
with open('./data/2.txt','w') as file:
    file.writelines(['Hello...\n', 'Hai...\n', "Again..."])
```

```python
with open('./data/3.txt','a') as file:
    file.writelines(['Hello...\n', 'Hai...\n', "Again...\n"])
```

```python
class MyContextManager:
    def __init__(self, name: str):
        self._name = name
    
    def __enter__(self):
        print(f"Welcome {self._name}")
        return self

    def visit(self):
        print("visiting")

    def __exit__(self, *args):
        print(f"Bye {self._name}")
```

```python
with MyContextManager('qt') as ctx:
    ctx.visit()
```

```python
# Generally it is good practice to use encodings
with open('./data/1.txt', mode="r", encoding="utf-8") as file:
    print(file.readlines())
```
