```python
movie_details = {
    "name": "RRR",
    "year": 2022,
    "cast": ["A", "B", "C"],
    "boxoffice": 1000.0
}
```

```python
import json
json_string = json.dumps(movie_details)
print(json_string)
```

```python
# writes to a json
with open("movie.json", "w") as f:
    json.dump(movie_details,f, indent=4)
```
