```python
percent = 58

if 93 <= percent <= 100:
    print("A")
elif 90 <= percent <= 92:
    print("A-")
elif 87 <= percent <= 89:
    print("B+")
elif 83 <= percent <= 86:
    print("B")
elif 80 <= percent <= 82:
    print("B-")
elif 77 <= percent <= 79:
    print("C+")
elif 73 <= percent <= 76:
    print("C")
elif 70 <= percent <= 72:
    print("C-")
elif 67 <= percent <= 69:
    print("D+")
elif 65 <= percent <= 66:
    print("D")
else:
    print("F")

```
