```python
def even_numbers(start:int=0, end:int=10) -> list[int]:
    even_numbers_list = list()
    while start <= end:
        if start % 2 == 0:
            even_numbers_list.append(start)
        start += 1
    return even_numbers_list
```

```python
even_numbers(start=20, end=60)
```

```python
def even_numbers_generator(start:int = 0, end:int = 0):
    while start <= end:
        if start % 2 == 0:
            yield start
        start += 1
```

```python
even_numbers = even_numbers_generator(10, 20)
next(even_numbers)
```

```python
def is_prime(number:int) -> bool:
    """Checks if the number is prime or not

    Args:
        number (int): number to be checked

    Returns:
        bool: Returns True if prime False otherwise
    """
    is_prime_result = True
    if number < 2:
        is_prime_result = False
    else:
        index = 2
        while index < number:
            if number % index == 0:
                is_prime_result = False
                break
            index += 1
    return is_prime_result
```

```python
start = 100
end = 1000
prime_numbers = []
for number in range(start, end+1):
    if is_prime(number):
        prime_numbers.append(number)
print(prime_numbers)
```

```python
prime_numbers = [ number for number in range(start, end+1) if is_prime(number) ]
print(prime_numbers)
```

```python
# euler 2
def fibb_sequence(max:int = 10):
    a, b = 0,1
    while a + b < max:
        c = a + b
        yield c
        a = b
        b = c

sum([number for number in fibb_sequence(4000000) if number%2 == 0])
```
