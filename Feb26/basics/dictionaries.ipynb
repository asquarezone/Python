# Dictionaries

* Keys are unique
* dictionaries can be iterated

```python
# loop

# inthe following style we get key

for movie in movie_director_dict:
    # key
    print(movie)
    # value
    print(movie_director_dict[movie])
```

```python
# in the following style we get both key and value
for movie, director in movie_director_dict.items():
    print("Movie : ", movie)
    print("Director :", director)
```

```python
list(movie_director_dict.keys())
```

```python
list(movie_director_dict.values())
```
