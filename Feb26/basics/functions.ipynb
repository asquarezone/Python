```python
# defining a function
def print_message(message:str, count:int)->None:
    """This functions prints the message multiple times
    depending on count

    Args:
        message (str): message to be printed
        count (int): number of times message should be printed
    """
    index = 0
    while index < count:
        print(message)
        index += 1
```

```python
def is_even(number:int) -> bool:
    """This function checks if the number is even or not

    Args:
        number (int): number to be checked

    Returns:
        bool: Returns True if even False otherwise
    """
    return number % 2 == 0
```

```python
def add(a: int|float, b: int|float) -> int|float:
    """Returns sum of two numbers

    Args:
        a (int|float):
        b (int|float):

    Returns:
        int|float: Return sum of two numbers
    """
    return a + b
```

```python
# passing arguments by position
add(5,6)
```

```python
# naming arguments
add(b=6, a=7)
```

```python
def power(number:int, p:int=1) -> int:
    """Calculates power

    Args:
        number (int): _description_
        p (int, optional): _description_. Defaults to 1.

    Returns:
        int: _description_
    """
    return number ** p
```

```python
power(10)
```

```python
power(5,2)
```

```python
power(number=10, p=3)
```
