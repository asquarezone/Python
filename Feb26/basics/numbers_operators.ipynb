```python
# calculate simple intrest
principal = 100000
interest_rate = 24
time_in_years = 5

simple_interest = principal * interest_rate * time_in_years / 100
print(simple_interest)

total_amount = principal + simple_interest
print(total_amount)
```

```python
principal = 100000
interest_rate = 24 / 100
time_in_years = 5
compound_frequency = 1

total_amount = principal * (1 + (interest_rate / compound_frequency) ) ** (compound_frequency * time_in_years)
total_amount
```
